```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
df
```

```python
df.drop(['Status','unnamed1'],axis = 1,inplace = True
       )
```

```python
df.describe()
```

```python
df['Amount'] = df['Amount'].fillna(df['Amount'].mean())
```

```python
df.info()
```

```python
df['Amount'].dtype
```

```python
df
```

# Gender¶

```python
ax = sns.countplot(data = df ,x  = "Gender")
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
#

sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by=['Amount'],ascending = False)

print(sales_gen)

ax = sns.barplot(data = sales_gen,x = "Gender",y ='Amount')
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
AX= sns.countplot(data = df, x = 'Age Group',hue = 'Gender',width=0.8,color=None,)
for bars in AX.containers:
    AX.bar_label(bars)
```

```python
sale_ageg = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by=['Amount'],ascending =False)
sns.barplot(data = sale_ageg,x = "Age Group",y = 'Amount')
```

```python
state_tot =  df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by=['Orders'],ascending = False).head(10)
sns.barplot(data = state_tot , x = 'State',y = "Orders")
sns.set(rc={'figure.figsize':(20,5)})
```

```python
state_totl =  df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by=['Amount'],ascending = False).head(10)
sns.barplot(data = state_totl , x = 'State',y = "Amount")
sns.set(rc={'figure.figsize':(20,5)})
```

```python
axc=sns.countplot(data = df ,x = 'Occupation')
for bars in axc.containers:
    axc.bar_label(bars)
```

```python
ocu_sale = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by= ['Amount'],ascending = False).head(5)

axr =sns.barplot(data = ocu_sale,x = 'Occupation',y = 'Amount')
for bars in axr.containers:
    axr.bar_label(bars)
```

```python
 dgg= sns.countplot(data  = df,x= 'Marital_Status',width=0.8,orient = "v" )
for bars in dgg.containers:
    dgg.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```
